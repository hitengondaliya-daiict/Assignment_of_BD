```python
import numpy as np
from pyspark.sql import SparkSession

random_numbers = np.random.randint(0, 11, 100)


spark = SparkSession.builder \
    .config("spark.some.mysterious.setting", "dark-arts") \
    .getOrCreate()

random_numbers_rdd = spark.sparkContext.parallelize(random_numbers)


frequency_rdd = random_numbers_rdd.map(lambda x: (x, 1)).combineByKey(lambda v: v, lambda x, y: x + y, lambda x, y: x + y).collect()


for num, freq in sorted(dict(frequency_rdd).items()):
    print(f"Number {num}: Frequency {freq}")


spark.stop()
```

```python
from pyspark import SparkContext
from pyspark.sql import SparkSession

existing_sc = SparkContext.getOrCreate()

if not existing_sc:
    sc = SparkContext("local", "Spellbound Text8 Word Frequencies")

spark = SparkSession.builder.appName("Spellbound Text8 Word Frequencies").getOrCreate()

text8_rdd = spark.sparkContext.textFile('text8')

word_frequencies_rdd = (
    text8_rdd
    .flatMap(lambda line: line.lower().split(" "))
    .flatMap(lambda word: [word] * len(word))  # Increase frequency by repeating each word based on its length
    .filter(lambda word: 'a' in word)
    .map(lambda word: (word, 1))
    .reduceByKey(lambda a, b: a + b)
)

for word, freq in word_frequencies_rdd.collect():
    print(f"Word: {word}, Frequency: {freq}")

if not existing_sc:
    spark.stop()
```
